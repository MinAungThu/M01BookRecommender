```python
import pandas as pd 
import numpy as np 
```

```python
book_path = "/Users/minaungthu/Documents/KU /1st year/2nd Semester/computer programming/project/dataset/Books.csv" 
user_path = "/Users/minaungthu/Documents/KU /1st year/2nd Semester/computer programming/project/dataset/Users.csv"
rating_path = "/Users/minaungthu/Documents/KU /1st year/2nd Semester/computer programming/project/dataset/Ratings.csv"

book_df = pd.read_csv(book_path)
user_df = pd.read_csv(user_path)
rating_df = pd.read_csv(rating_path)
```

```python
book_df.columns
```

```python
book_df = book_df[['ISBN','Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-L']]
```

```python
book_df.columns
```

```python
user_df.columns
user_df.rename(columns={"User-ID":'user_id',"Location":"location", "Age":'age'},inplace=True)
user_df.columns
```

```python
rating_df.columns
```

```python
rating_df.rename(columns={"User-ID":"user_id", "Book-Rating":"rating"},inplace=True)
rating_df.columns
```

```python
print(f"book_df size: {book_df.shape} \nuser_df size : {user_df.shape} \nrating_df size: {rating_df.shape}")
```

```python
#storing users who rated at least 100 books 
x = rating_df['user_id'].value_counts() > 100
```

```python
x[x].shape
```

```python
y= x[x].index
y
```

```python
rating_df = rating_df[rating_df['user_id'].isin(y)]
rating_df.head()
```

```python
rating_df.shape
```

```python
books_with_ratings = rating_df.merge(book_df, on='ISBN')
books_with_ratings.head()
```

```python
number_rating = books_with_ratings.groupby('Book-Title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = books_with_ratings.merge(number_rating, on='Book-Title')
final_rating.head()
```

```python
#taking books with at least 30 ratings 
final_rating = final_rating[final_rating['num_of_rating'] >= 30]
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','Book-Title'],inplace=True)
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='Book-Title', values= 'rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
#since the dataset is mostly sparse, we will apply sparse matrix 
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
type(book_sparse)
```

```python
!pip install scikit-learn
```

```python
from sklearn.neighbors import NearestNeighbors
#using brute force since the input data is very sparse
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[1,:].values.reshape(1,-1), n_neighbors=6 )
```

```python

for i in suggestion[0]:
    print(book_pivot.index[i])
```

```python

def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggested books are: \n")
                else:
                    print(j)
```
